```
#import libraries
import pandas as pd
import numpy as np
```

```
#loading datasets

telcom1 = pd.read_csv('Telcom_dataset.csv')
telcom1.head(10)



a = ['PRODUCT','VALUE','DATE_TIME','CELL_ON_SITE',
'DW_A_NUMBER_INT',
'DW_B_NUMBER_INT',
'COUNTRY_A',
'COUNTRY_B',
'CELL_ID',
'SITE_ID']


telcom1.columns.values[:] = [i for i in a ]

telcom1
```

```
#Getting rid of unwanted columns for the first data set
# uwanted_columns = ['COUNTRY_A','COUNTRY_B','CELL_ON_SITE']

telcom_1_clean = telcom1.drop(columns=(['COUNTRY_A','COUNTRY_B','CELL_ON_SITE']))
telcom_1_clean
```

```
#reading and renaming columns for the second dataset provided

telcom2 = pd.read_csv('Telcom_dataset2.csv')
telcom2.head(10)

telcom2.columns = ['PRODUCT','VALUE','DATE_TIME','CELL_ON_SITE',
'DW_A_NUMBER_INT',
'DW_B_NUMBER_INT',
'COUNTRY_A',
'COUNTRY_B',
'CELL_ID',
'SITE_ID']
```

```
#Droping the uncessary columns forthe second dataset

telcom_2_clean = telcom2.drop(columns=(['COUNTRY_A','COUNTRY_B','CELL_ON_SITE']))
telcom_2_clean
```

```
#read the third dataset
telcom3 = pd.read_csv('Telcom_dataset3.csv')
telcom3.head(10)

telcom3.columns = ['PRODUCT','VALUE','DATE_TIME','CELL_ON_SITE',
'DW_A_NUMBER_INT',
'DW_B_NUMBER_INT',
'COUNTRY_A',
'COUNTRY_B',
'CELL_ID',
'SITE_ID']
```

```
#Drop the unwanted columns)

telcom_3_clean = telcom3.drop(columns=(['COUNTRY_A','COUNTRY_B','CELL_ON_SITE']))
telcom_3_clean
```

```
# columns=['VILLES',
# 'STATUS',
# 'LOCALISATION',
# 'DECOUPZONE',
# 'ZONENAME',
# 'LONGITUDE',
# 'LATITUDE',
# 'REGION',
# 'AREA',
# 'CELL_ID',
# 'SITE_CODE']
deli = [',',' ',',']
cells = pd.read_csv('/content/cells_geo.csv', delimiter=';')
cells_df = pd.DataFrame(cells)
cells_df.head(10)
```

```
#CONCATNATE ALL THE TABLES TO BE ABLE TO DO SOME OPERATIONS ON THE DATA

dt1 = pd.concat([telcom_1_clean,telcom_2_clean,telcom_3_clean],axis = 0,sort = False)
dt1
```

```
#check for null values and delete rows with no data and value is equal to 0

dt1.dropna(how='all')
dt1 = dt1[dt1.VALUE != 0]
dt1
dt1 = dt1.drop_duplicates()
dt1
```

```
#replacing null values with 0 and converting it to lowercase
dt1 = dt1.fillna('.')
dt1 = dt1.astype(str).apply(lambda x: x.str.lower())
dt1
```

```
#conveting the geo dataset to lowercase


# cells_df = cells_df.fillna(0)

# cells_df = cells_df.dropna()
cells_df = cells_df.astype(str).apply(lambda x: x.str.lower())
cells_df
cells_df = cells_df.drop_duplicates()
cells_df.describe()
```

```
# Which ones were the most used city for the three days?
#We merger our dataset with the geographical data on the common column which is cell_ID where
combined = dt1.merge(cells_df, left_on= 'SITE_ID', right_on='CELL_ID')
combined.head(10)

combined = combined.drop_duplicates()
combined.head(10)
```

```
# merged = dt1.merge(cells_df, how='inner', on='CELL_ID')
# most_city = merged.groupby('VILLES')['VALUE'].sum()
# most_city.sort_values(ascending=False)
# # merged
# most_city

combined_count = combined.groupby('VILLES')['VILLES'].count()
combined_count.sort_values(ascending=False)
```

```
# Which cities were the most used during business and home hours?
#split date and time columns

dt1_no_date = dt1.drop(columns='DATE_TIME')
dt1_no_date

#time column
dt2 = dt1['DATE_TIME'].apply(lambda x: pd.Series([i for i in reversed(x.lower().split(' '))]))
dt2.columns =['DATE','TIME']
dt2_time = dt2.drop(columns='DATE')
dt2_time

#date column
dt2 = dt2['DATE'].apply(lambda x: pd.Series([i for i in reversed(x.lower().split('.'))]))
dt2.columns = ['index', 'DATE']
dt3_date = dt2.drop('index',axis=1)
dt3_date


dt4 = pd.concat([dt1_no_date, dt2_time, dt3_date],axis = 1,sort = False)
dt4.head()
dt4.rename(columns= {'TIME':'DATE','DATE':'TIME'},inplace =True)
dt4

#using the cleaned date/time dateset and the geo data we can find the cities 

business_hours = (dt4['TIME'] < '08:00:00') 
business_hours.describe()
```
